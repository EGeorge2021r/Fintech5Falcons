###   - Create new column in the estate data and named starbucks_exist

###   - Create a list of zip codes from starbucks data and call it starbucks_Zipcodes from clean_starbucks_opening_df

```python
# Generate a list of all zipcodes from the starbuck dataframe
zip_List = list(clean_austin_starbucks_df['zip_code'])
#zip_List
```

## Step 2: Analyze the data trend to establish location of Starbucks 
#


```python
# Create a function to search for a zipcode in the zipcode list and return result
# If ItemToReturn is "Exist", then it will return 1 for exist or 0 for non-exist
# If ItemToReturn is "Count", then it will return the count of zipcode found in the list
# Else, it will return -1 for error data

def starbucks_ExistCount(zip_code,zip_List,ItemToReturn):
    zip_exist = 0
    zip_count = zip_List.count(str(zip_code))
    if zip_count > 0:
        zip_exist = '1'
    else:
        zip_exist = '0'
    if ItemToReturn == 'Count': return(zip_count)
    if ItemToReturn == 'Exist': return(zip_exist)
    return(-1)
```

```python
# Create new columns in the Austin  estate DataFrame and calculate its value from the function (Starbuck_ExistCount) above
austin_estate_df['StarBucksExist'] = austin_estate_df.apply(lambda x: starbucks_ExistCount(x['zipCode'],zip_List,"Exist"), axis=1)
austin_estate_df['StarBucksCount'] = austin_estate_df.apply(lambda x: starbucks_ExistCount(x['zipCode'],zip_List,"Count"), axis =1)
```

```python
# Add Starbucks exist and starbucks count data to the clean_austin_starbucks_df to be used for geo-spatial plot
combined_df = clean_austin_starbucks_df
combined_df[['ZipCode_Exist','ZipCode_Count']]=0
# Review the DataFrame
combined_df
```

```python
# Create a function to search for a zipcode in the zipcode list and return result
# If ItemToReturn is "Exist", then it will return 1 for exist or 0 for non-exist
# If ItemToReturn is "Count", then it will return the count of zipcode found in the list
# Else, it will return -1 for error data

def Estate_ExistCount(zipCode,zip_List2,ItemToReturn):
    zip_exist2 = 0
    zipCode=int(zipCode)
    zip_count2 = zip_List2.count(zipCode)
    if zip_count2 > 0:
        zip_exist2 = 'Y'
    else:
        zip_exist2 = 'N'
    if ItemToReturn == 'Count2': return(zip_count2)
    if ItemToReturn == 'Exist2': return(zip_exist2)
    return(-1)
```

```python
# Create new columns in the starbucks DataFrame and calculate its value from the function (Estate_ExistCount) above
combined_df['ZipCode_Exist'] = combined_df.apply(lambda x: Estate_ExistCount(x['zip_code'],zip_List2,"Exist2"), axis=1)
combined_df['ZipCode_Count'] = combined_df.apply(lambda x: Estate_ExistCount(x['zip_code'],zip_List2,"Count2"), axis =1)
combined_df
```

```python
# Create a plot to analyze Estate home values in relations to Starbucks location info
# Show the plot
combined_df.hvplot.points(
    'longitude', 
    'latitude', 
    geo=True, 
    color='zip_code',
    size ='ZipCode_Count',
    scale=2.0,
    tiles='OSM',
    title='Austin Starbucks geo-locations',
    frame_width=700,
    frame_height=500
)
```
