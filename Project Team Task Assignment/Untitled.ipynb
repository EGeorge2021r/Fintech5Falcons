```python
# The API Call endpoint URLs for the locations and estates are:
starbucks_url = "http://www.starbuckseverywhere.net/StoreOpeningDates.html"
#starbucks_ur2 = "https://www.google.com/maps/search/starbucks+in+Austin,+TX/@30.2990239,-97.7999839,11z"
estate_url = "https://realty-in-us.p.rapidapi.com/locations/auto-complete"
yelp_url = "https://yelpapiserg-osipchukv1.p.rapidapi.com/getAutocomplete"
```

```python
# Using the Beautiful soup in Python requests library, make an API call to access the starbucks location html.
# BeautifulSoup is a Python library for parsing HTML and XML documents

#path='http://www.starbuckseverywhere.net/StoreOpeningDates.html'
path = starbucks_url
# empty list
data = []

# for getting the header from the HTML file

list_header = ["Starbucks opening dates"]
soup = BeautifulSoup(open(path),'html.parser')
header = soup.find_all("table")[0].find("tr")
  
for items in header:
    try:
         list_header.append(items.get_text())
    except:
        continue
            
#for getting the data 
HTML_data = soup.find_all("table")[0].find_all("tr")[1:]
  
for element in HTML_data:
     sub_data = []
for sub_element in element:
    try:
            sub_data.append(sub_element.get_text())
    except:
             continue
data.append(sub_data)

# Storing the data into Pandas DataFrame 
dataFrame = pd.DataFrame(data = data, columns = list_header)

# Converting Pandas DataFrame into CSV file
dataFrame.to_csv('starbucks.csv')
```
